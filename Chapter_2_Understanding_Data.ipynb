## **Distance**
- จุดแต่ละจุดต่างกันเเค่ไหน

```python
data = [[1,2],[-1,4],[0,3],[-3,-2]]
df = pd.DataFrame(data,columns=['X1','X2']) #ไม่ตั้งชื่อเป็น x,y คนละความหมาย
df
```

```python
# prompt: draw scatter plot of the df, draw line for X1 and X2 axes

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 'X1' and 'X2' columns
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'])
plt.plot([-3,1],[0,0], color='black',linewidth=2)  # X2 axis (horizontal line at y=0)
plt.plot([0,0],[-2,4], color='black' ,linewidth=2)  # X1 axis (vertical line at x=0)


# Customize the plot (optional)
plt.xlabel('X1',fontweight='bold' )
plt.ylabel('X2',fontweight='bold')
plt.title("Scatter Plot of X1 vs X2")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: calulate distatance matrix of the df

import pandas as pd
from scipy.spatial.distance import pdist, squareform

# Assuming 'df' is your DataFrame with columns representing coordinates
# Example: df[['X1', 'X2']] if your coordinates are in columns 'X1' and 'X2'
coordinates = df[['X1', 'X2']]

# Calculate the pairwise distances
distance_matrix = squareform(pdist(coordinates))
distance_matrix
```
